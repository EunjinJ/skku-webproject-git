```python
import requests
import pandas as pd 
from bs4 import BeautifulSoup 

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from tqdm import tqdm
import time
```

```python
s = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
```

```python
kakao_jongro_cafe_link = pd.read_csv('C:/Users/user/Desktop/project_5/crawl_data_map/kakao_jongro_cafe_link.txt').drop(columns=['Unnamed: 0'], axis=1)
kakao_jongro_cafe_link
```

```python
url = kakao_jongro_cafe_link['place_link'][451]
driver.get(url)
```

```python
# 이미지
driver.find_element(By.CLASS_NAME, 'bg_present').get_attribute('style').split('"')[1]
```

```python
driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > h2').text
```

```python
place_img_data = {
    'place_name':[],
    'place_img' : []
}
```

```python
for i in tqdm(range(453, len(kakao_jongro_cafe_link))):
    url = kakao_jongro_cafe_link['place_link'][i]
    driver.get(url)

    time.sleep(2)
    
    # 장소 이름
    place_img_data['place_name'].append(driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > h2').text)

    # 장소 상세 정보    
    if driver.find_elements(By.CLASS_NAME, 'bg_present') == []:
        place_img_data['place_img'].append('none')
    else:
        place_img_data['place_img'].append(driver.find_element(By.CLASS_NAME, 'bg_present').get_attribute('style').split('"')[1])
        
```

```python
kakao_jonro_cafe_img = pd.DataFrame(place_img_data).drop_duplicates()
kakao_jonro_cafe_img
```

```python
kakao_jonro_cafe_img['place_img'][6]
```

```python
kakao_jongro_cafe_img = pd.DataFrame(place_img_data).drop_duplicates()
kakao_jongro_cafe_img.to_csv('kakao_jongro_cafe_img')
kakao_jongro_cafe_img
```
