```python
s = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=s)
```

```python
def crawl_link(key):
    url = 'https://map.kakao.com/'
    driver.get(url)
    driver.refresh()
    
    # 키워드 입력
    driver.find_element(By.CSS_SELECTOR, '#search\.keyword\.query').send_keys(key)
    driver.find_element(By.CSS_SELECTOR, '#search\.keyword\.submit').click()

    driver.implicitly_wait(100)

    # 장소 더보기
    driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.more').click()
```

```python
place_link_data = {
    'place_name':[],
    'place_link':[],
    'place_page':[]
}
```

```python
for i in range(len(driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a'))):
    driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a')[i].click()
    time.sleep(2)
    
    page_place = driver.find_elements(By.CSS_SELECTOR, '#info\.search\.place\.list > li')

    for y in range(len(page_place)):
        place_link_data['place_name'].append(page_place[y].find_element(By.CSS_SELECTOR, 'div.head_item.clickArea > strong > a.link_name').text)

        place_detail = page_place[y].find_elements(By.CSS_SELECTOR, 'div.info_item > div.contact.clickArea > a')
        
        place_link_data['place_link'].append(place_detail[0].get_attribute('href'))
        
        if place_detail[1].get_attribute('href') == 'https://map.kakao.com/#none':
            place_link_data['place_page'].append('none')
        else:
            place_link_data['place_page'].append(place_detail[1].get_attribute('href'))

driver.find_element(By.CSS_SELECTOR, '#info\.search\.page\.next').click()
time.sleep(2)        
```

```python
place_info_detail = {
    'place_name' : [],
    'place_category_detail' : [],
    'place_adress' : [],
    'place_running_date' : [],
    'place_page' : [],
    'place_phone' : [],
    'place_tag' : [],
    'place_info' : [],
}
```

```python
for i in tqdm(range(len(kakao_jonro_search_link_data))):
    url = kakao_jonro_search_link_data['place_link'][i]
    driver.get(url)

    time.sleep(2)
    
    # 장소 이름
    place_info_detail['place_name'].append(driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > h2').text)
    # 장소 카테고리 상세
    place_info_detail['place_category_detail'].append(driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > div > span.txt_location').text)
    
    place_table = driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div.details_placeinfo')
    
    # 장소 주소
    place_info_detail['place_adress'].append(place_table.find_element(By.CLASS_NAME, 'txt_address').text)
    
    # 장소 운영시간
    if place_table.find_elements(By.CLASS_NAME, 'txt_operation') == []:
        place_info_detail['place_running_date'].append('None')
    else:
        place_info_detail['place_running_date'].append(place_table.find_element(By.CLASS_NAME, 'txt_operation').text)
    
    # 장소 상세 페이지
    if place_table.find_elements(By.CLASS_NAME, 'link_homepage') == []:
        place_info_detail['place_page'].append('None')
    else:
        place_info_detail['place_page'].append(place_table.find_element(By.CLASS_NAME, 'link_homepage').text)

    # 장소 전화번호    
    if place_table.find_elements(By.CLASS_NAME, 'txt_contact') == []:
        place_info_detail['place_phone'].append('None')
    else:
        place_info_detail['place_phone'].append(place_table.find_element(By.CLASS_NAME, 'txt_contact').text)

    # 장소 태그    
    if place_table.find_elements(By.CLASS_NAME, 'txt_tag') == []:
        place_info_detail['place_tag'].append('None')
    else:
        place_info_detail['place_tag'].append(place_table.find_element(By.CLASS_NAME, 'txt_tag').text)

    # 장소 상세 정보    
    if place_table.find_elements(By.CLASS_NAME, 'list_facility') == []:
        place_info_detail['place_info'].append('none')
    else:
        place_info_detail['place_info'].append(', '.join(place_table.find_element(By.CLASS_NAME, 'list_facility').text.split('\n')))
        
```
